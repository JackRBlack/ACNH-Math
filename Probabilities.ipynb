# Probabilities

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Lily of the valley

```python
day = np.arange(1, 11, 1)
p = [(x**2 + 6)/100 for x in day]
p[-1] = 1

p_success = [p[0]]
p_fail = 1
for i in range(1, 10):
    p_fail = p_fail * (1 - p[i-1])
    p_success.append(p_fail * p[i])
```

```python
sns.set_theme(style="whitegrid")
```

```python
fig = plt.figure(figsize=(10, 4), dpi=200)

fig.add_subplot(121)
ax = sns.barplot(x=day, y=p_success, color='C2')
plt.yticks(np.arange(0, 0.25, 0.05))
plt.xlabel("Day")
plt.ylabel("Probability")
plt.title("Bloom at Day x")

fig.add_subplot(122)
ax = sns.barplot(x=day, y=p, color='C3')
plt.yticks(np.arange(0, 1.2, 0.2))
plt.ylim(0, 1)
plt.xlabel("Day")
plt.ylabel("Probability")
plt.title("Bloom at Day x If NOT Blooming Before")

plt.tight_layout()
plt.savefig("./Lily-of-the-valley_probability.png")
plt.show()
```
